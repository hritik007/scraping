```python
import requests #to request to the web site from where you want to do scrapping
import bs4 #for beautiful soup
```

```python
#res variable is holding the request (res variable will store entire webpage)
res = requests.get('https://www.ratestar.in/company/540697/AandM-Febcon-Ltd-286682')
```

```python
soup = bs4.BeautifulSoup (res.text, 'lxml')
print(soup.prettify())
```

```python
#find all html source of tables
temp=soup.find_all('tr')

#to make list of tables in a html
all_tables = []

for i,t in enumerate(temp):
    print(".......................table["+str(i)+"]..............................")
    print (t.getText)
    print(".............................end.................................")
    print()
    all_tables.append(t)


#temp[0].getText() #for quaterly report
#a=temp[1].getText() #for annual report
#temp[2].getText() #for balance sheet
#for b in a:
    #print(b)
```

```python
#to make list of words in a given table

words = []

temp = all_tables[1].getText()
splits = temp.split()
for i,t in enumerate(splits):
    print(str(i)+")  "+str(t))
    words.append(t)
```
