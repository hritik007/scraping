```python
#removing unwanted spaces, tbas,'Notes' etc
all_rows_new = []
remove_next_flag = False

for i in range(len(all_rows)):
    temp = all_rows[i]
    temp = temp.replace('\n','')
    temp = temp.replace('  ','')
    temp = temp.replace('\xa0',' ')

    if(not(temp == "")):
        if(remove_next_flag == True):
            remove_next_flag = False
        elif(temp[-1] == ":"):
            remove_next_flag = True
            continue
        elif(temp == 'Notes'):
            all_rows_new.append('Notes')
        else:
            all_rows_new.append(temp)
    else:
        all_rows_new.append("empty")
    
print(len(all_rows_new))
all_rows_new
```

```python
row_names = []

for r in all_rows_new:
    if r[0].isdigit() == False and (not (r[0] == "-") and (not r == 'empty')): #if row name comes
        row_names.append(r)
row_names
```

```python
def return_row(all_rows_new):
    temp_row = []
    for r in all_rows_new:
        if r[0].isdigit() == False and (not (r[0] == "-")): #if row name comes
            temp_row = []
        #also assign row name as index name of dataframe
        #print(r)
        else: #if row value comes
            temp_row.append(r)

    return temp_row
```

```python
#splitting all_rows_new into 5 parts (coz 5 tables)
table1 = []
table2 = []
table3 = []
table4 = []
table5 = []

counter = 0
for r in all_rows_new:
    if r == 'Sales +' or r == 'Share Capital +' or r == 'Cash from Operating Activity +' or r == 'ROCE %':
        counter = counter + 1
    
    if counter == 1:
        table1.append(r)
    if counter == 2:
        table2.append(r)
    if counter == 3:
        table3.append(r)
    if counter == 4:
        table4.append(r)
    if counter == 5:
        table5.append(r)
        
#dropping 12 Notes and 1 blank index name
rangee = columns_in_each_table['Quarters']
for i in range(rangee+1):
    table1.pop()
```
